```python
from sklearn.ensemble import RandomForestClassifier;

from sklearn.model_selection import GridSearchCV;

param_grid = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt'], "bootstrap" : [True],
              'n_estimators':[10, 20, 50, 80, 100], 
               "max_samples":[1.0]} 

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3, n_jobs=-1, scoring='accuracy',error_score='raise') 


grid.fit(X_ROS, y_ROS);
```

```python
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold
from sklearn import linear_model
from sklearn import datasets
```

```python
# Create StratifiedKFold object.
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
lst_accu_stratified = []
```

```python
RC=RandomForestClassifier()
for i,(train_index, test_index) in enumerate(skf.split(X_ROS, y_ROS)):
    x_train_fold, x_test_fold = X_ROS.iloc[train_index], X_ROS.iloc[test_index]
    y_train_fold, y_test_fold = y_ROS[train_index], y_ROS[test_index]
    RC.fit(x_train_fold, y_train_fold)
    from sklearn.metrics import confusion_matrix
    y_pred=RC.predict(x_test_fold)
    RFConfusionMatrix = confusion_matrix(y_test_fold, y_pred)
    print("Split",i)
    print(RFConfusionMatrix)
    from sklearn.metrics import classification_report
    RFClassifierReport = classification_report(y_test_fold, y_pred)
    print(RFClassifierReport)

    lst_accu_stratified.append(RC.score(x_test_fold, y_test_fold))
  #Print the output.
print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(lst_accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
      min(lst_accu_stratified)*100, '%')
print('\nOverall Accuracy:',
      mean(lst_accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```
